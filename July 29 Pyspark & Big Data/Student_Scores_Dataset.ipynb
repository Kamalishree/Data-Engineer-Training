```python
import pandas as pd

# Read from Google Drive path
file_path = '/content/drive/MyDrive/students.csv'
df = pd.read_csv(file_path)

# Add 'Average' column
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

# Display the updated DataFrame
print(df)

# Print student with the highest average
top_student = df.loc[df['Average'].idxmax()]
print("\nStudent with the highest average:\n")
print(top_student)
```

```python
# Students who scored more than 80 in Science
high_science = df[df['Science'] > 80]
print("Students who scored more than 80 in Science:\n")
print(high_science)

# Students with average score below 70
low_avg = df[df['Average'] < 70]
print("\nStudents with average score below 70:\n")
print(low_avg)
```

```python
# Sort by English score in descending order and print top 3 students
top_english = df.sort_values(by='English', ascending=False).head(3)

print("Top 3 students by English score (descending):\n")
print(top_english)
```

```python
import matplotlib.pyplot as plt

# Bar chart: Name vs Average
plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Average'], color='skyblue')
plt.title('Average Score per Student')
plt.xlabel('Student Name')
plt.ylabel('Average Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Line chart for Math, Science, and English
plt.figure(figsize=(10, 6))
for subject in ['Math', 'Science', 'English']:
    plt.plot(df['Name'], df[subject], marker='o', label=subject)

plt.title('Student Scores by Subject')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/students.csv')
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)
```

```python
# Calculate total for each subject
total_scores = {
    'Math': df['Math'].sum(),
    'Science': df['Science'].sum(),
    'English': df['English'].sum()
}

# Find subject with highest total
highest_subject = max(total_scores, key=total_scores.get)

print("Total Scores per Subject:")
print(total_scores)

print(f"\n Subject with the Highest Total Score: {highest_subject} ({total_scores[highest_subject]})")

# Save updated DataFrame (with Average column) back to Google Drive

output_path = '/content/drive/MyDrive/students_summary.csv'
df.to_csv(output_path, index=False)


```
